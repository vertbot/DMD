```python
import pandas as pd
xls = pd.ExcelFile('DMD Majors as of 01.07.2022.xlsx')


pd.set_option('display.max_rows', 2000)

df1 = pd.read_excel(xls, 'Storrs')
df2 = pd.read_excel(xls, 'Stamford')

df3 = pd.read_excel(xls, 'Alumni')
frames = [df1, df2, df3]

full_df = pd.concat(frames)
```

```python

full_df.rename(columns = {"Grad Year": "Grad_Year", "Notes/ Current Employer": "Notes_Current_Employer", "personal email": "personal_email", "Address 2": "Address_2"}, inplace = True)

full_df['NetId'] = full_df['NetId'].astype(str)
full_df['Peoplesoft'] = full_df['Peoplesoft'].astype(int)
full_df['First'] = full_df['First'].astype(str)
full_df['Last'] = full_df['Last'].astype(str)
full_df['Degree'] = full_df['Degree'].astype(str)
full_df['Concentration'] = full_df['Concentration'].astype(str)
full_df['Campus'] = full_df['Campus'].astype(str)
full_df['Advisor'] = full_df['Advisor'].astype(str)
full_df['Degree'] = full_df['Degree'].astype(str)
full_df['Notes_Current_Employer'] = full_df['Notes_Current_Employer'].astype(str)
full_df['Email'] = full_df['Email'].astype(str)
full_df['personal_email'] = full_df['personal_email'].astype(str)
full_df['Address_2'] = full_df['Address_2'].astype(str)
full_df['City'] = full_df['City'].astype(str)
full_df['State'] = full_df['State'].astype(str)
full_df['Zip'] = full_df['Zip'].astype(str)
full_df['Phone'] = full_df['Phone'].astype(str)
full_df['Zip'] = full_df['Zip'].astype(str)
full_df['Grad_Year'] = full_df['Grad_Year'].astype(str)

full_df.dropna(how='all', axis=1, inplace=True)




#full_df.columns
```
